
### Insights on Movie Duration (in Minutes)

The filtered dataset provides a detailed look at movie durations, with the following insights:

- **Most Common Durations**: 
  - Movies with durations around **90 minutes** (e.g., **90 min**, **94 min**, **93 min**, **97 min**) are the most common, each appearing between **146-152** times.
  - These durations are typical for many feature films, which often fall within this range.
  
- **Less Common Durations**: 
  - **212 min**, **186 min**, **193 min**, and similar values appear only **once** in the dataset, suggesting the presence of very long films or special editions.
  - Short films or special features may explain the occurrence of **8 minutes**, which is far less common.

- **Trends in Film Length**:
  - The dominance of **90 minutes** suggests that a large proportion of the content is made up of traditional-length movies.
  - The spread of longer and shorter durations highlights the variety of content available, from full-length films to shorter TV specials or documentaries.

### Conclusion:
The data reveals clear preferences for specific movie lengths (primarily in the 90-100 minute range), while also showing a minority of longer films and rare short-form content.

```python
type_data = data[['type', 'release_year']]

tv_shows = type_data[type_data['type'] == 'TV Show'].groupby('release_year')['type'].count()
movies = type_data[type_data['type'] == 'Movie'].groupby('release_year')['type'].count()


plt.figure(figsize=(15, 8))  # Set the figure size

plt.plot(tv_shows.index, tv_shows.values, label='TV show', color='white', linewidth=2, markersize=6)


plt.plot(movies.index, movies.values, label='Movie', color='red', linewidth=2, markersize=6)


plt.title('Movie vs TV Show, Trend Analysis per year_added', fontsize=20, color='white')
plt.xlabel('Year', fontsize=14, color='black')
plt.ylabel('Count', fontsize=14, color='black')


plt.legend(fontsize=14, loc='upper left', facecolor='white', edgecolor='white')
plt.grid(color='black', linestyle='--', linewidth=0.5)
plt.gca().set_facecolor('#4f4f4f')  # Set background to black
plt.xticks(fontsize=12, color='grey')
plt.yticks(fontsize=12, color='black')


plt.text(2017, 600, 'Movie', fontsize=14, color='red')
plt.text(2017, 450, 'TV show', fontsize=14, color='white')
plt.text(1940, 700, 'Movie VS TV show analysis trend per release year', fontsize=20, color='white')

plt.show()
```
